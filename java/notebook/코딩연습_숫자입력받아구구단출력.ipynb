```java
import java.lang.*;
import java.io.*;
```

```java
public class solution {
    
    int from;
    int to;
    
    public solution (int x, int y) {
        
        if(x >= y) {
            
            this.from = y;
            this.to = x;
        
        }
        else if(x < y) {
            
            this.from = x;
            this.to = y;
        
        }
}
    
    public void showGugudan() {

        for(int i=this.from; i<=this.to; i++) {
            
            for(int j=1; j<10; j++) {
                
                System.out.println(i+"X"+j+"="+i*j);
                if(j == 9){
                    System.out.println("");
                
                }

            }
        }
    }
    
}
```

```java
//숫자 입력 하여 배열에 저장하기
System.out.println("숫자 2개를 입력하세요!");

String[] str = new String[2];
int x[] = new int[2];
BufferedReader br = new BufferedReader(new InputStreamReader(System.in));

for(int i=0; i<=1; i++) {
    str[i] = br.readLine();
    x[i] = Integer.parseInt(str[i]);
}
```

```java
solution s = new solution(x[0], x[1])
```

```java
s.showGugudan()
```

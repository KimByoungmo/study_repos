# Welcome to Kite's JupyterLab Extension Tutorial üéâ

Kite is your programming copilot. Kite will try to show you the right information at the right time as you code to prevent you from context switching out of your current line of thought.

## This is an _interactive_ tutorial üëã

You'll try out all of Kite's core features while working with a sample data set. You should be able to learn everything in 5 minutes. 

If you get stuck at any point, please visit [our help center](https://help.kite.com/) or [file an issue](https://github.com/kiteco/issue-tracker).

## Before we start

Kite's JupyterLab extension will by default try to start the Kite backend when the editor first starts. 

The Kite icon ü™Å in the bottom left corner of JupyterLab's status bar tells you if Kite is ready and working. 
- `ready` or `indexing` means you're good to go. Proceed to the next step!
- `not running` means you need to start the Kite Engine. Open the command palette `Cmd/Ctrl + Shift + C` and use the command `Kite: Open Copilot`
- `not installed` means you need to install Kite with [these instructions](https://github.com/kiteco/jupyterlab-kite/#installation).

# Part 1: Code Completions

Kite analyzes your code and uses machine learning to show you completions for what you're going to type next. It is able to understand dynamic libraries to give you smart completions.

![Kite vs Other](https://kite.com/kite-public/with_vs_without_small.png)

## Name completions

Kite can suggest names of symbols to use, such as names of packages or names of variables in scope.

![image](https://kite.com/kite-public/import_statement.png)

### Let's try it out!

- Put your cursor at the end of the line marked with `<--`
- Type `a` and select the completion for `pandas as pd` Here, [Kite Pro](https://www.kite.com/pro/) recommended a common alias for you.

```python
import p    # <-- import pandas as pd
```

## Built-in Completions, now as you type

Jupyter's built-in completions previously only showed up on `tab`. Now, they are available automatically as you type.

![Kernel completions](https://kite.com/kite-public/kernel.png)

### Let's try it out!

- Put your cursor at the end line of the line marked with `<--`
- Type `e` and select the completion for `sepal_length`

```python
df.s    # <-- df.sepal_length
```

## Code completions on demand

Remember that you can press `tab` at any time to request code completions.

### Let's try it out!

- Put your cursor at the end line of the line marked with `<--`
- Type `tab` and select the completion for `petal_width`

```python
df.pe    # <-- df.petal_width
```

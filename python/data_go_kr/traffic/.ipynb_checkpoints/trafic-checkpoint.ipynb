```python
mpl.rcParams["figure.figsize"] = (60, 10)
g = sns.FacetGrid(row="월", col = "호선", data = df_month, sharex = False, sharey = False)
g.map(plt.plot, "역명", "value", marker = "o")

for i in g.axes.flat :
    i.set_xticklabels(labels = i.get_xticklabels(), rotation = 90)

plt.show()
```

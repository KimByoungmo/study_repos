```python
from google.cloud import bigquery
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

client = bigquery.Client()
```

```python
ds_ref = client.dataset("usa_names", "bigquery-public-data")
ds = client.get_dataset(ds_ref)
```

```python
[i.table_id for i in client.list_tables(ds)]
```

```python
client.get_table(ds.table("usa_1910_2013")).schema
```

```python
sql = """SELECT YEAR,STATE,GENDER,NAME,SUM(NUMBER) AS NUM 
         FROM `bigquery-public-data.usa_names.usa_1910_2013`
         GROUP BY YEAR,STATE,GENDER,NAME
         """
```

```python
df_1910_2000 = client.query(sql).to_datefmame()
```

```python
for i,j in enumerate(df["Top Ten"]):
    if j is np.nan or j == "\n":
        df["Top Ten"].iloc[i] = ""
```

```python
df["Top Ten"].unique()
```

```python
df.head()
```

```python
year_result = []
rank_result = []
for i,j in enumerate(df["Top Ten"]) :
    if j == "" :
        year_result.append("")
        rank_result.append("")
    else :
        year_result.append(str(j).split(" ")[0])
        rank_result.append(str(j).split(" ")[1].replace("#",""))
```

```python
df["Top Ten_year"] = pd.Series(pd.to_numeric(year_result))
df["Top Ten_rank"] = pd.Series(pd.to_numeric(rank_result))
```

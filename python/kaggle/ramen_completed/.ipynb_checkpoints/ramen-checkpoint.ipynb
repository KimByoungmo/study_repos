```python
df.head()
```

```python
temp = df[(pd.isna(df["Top Ten"]) == False) & (df["Top Ten"]!= "\n")]["Top Ten"]
result = []
temp1 = []
for i, j in zip(temp.index.values, temp):
    temp1 = [i]
    temp1.extend(j.split(" "))
    result.append(temp1)
result = pd.DataFrame(result, columns=["index","year","rank"])
result = result[["year","rank"]].set_index(result["index"])
```

```python
df = df.join(result)
```

```python
plt.figure(figsize = (24,36))
sns.heatmap(df.pivot_table(index="Brand", columns="Style", values="Stars"))
```

```python
res = []
for i in df["rank"]:
    temp = str(i).replace("#","")
    res.append(temp)
df["rank"] = pd.Series(res).astype("float32")
```

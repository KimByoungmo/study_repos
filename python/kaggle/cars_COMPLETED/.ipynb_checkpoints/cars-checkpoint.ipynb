```python
df["condition"].head()
```

```python
result = []
for i in df["condition"] :
    if i.find("days") >=0  and i.find("hours") >= 0 :
        result.append(False)
    else :
        result.append(True)
all(result)
```

```python
import re
number = re.compile("[0-9]+")
res = []
for i in df["condition"] :
    if i.find("day") >= 0 :
        res.append((int(number.findall(i)[0])) * 24)
    elif i.find("hour") >= 0 :
        res.append(int(number.findall(i)[0]))
    elif i.find("minute") >= 0 :
        res.append(int(number.findall(i)[0]) / 60)
    else :
        res.append(0)
```

```python
df["hours"] = pd.Series(res)
```

```python
sns.catplot(x="brand", y="hours", hue="title_status", data=df)
plt.xticks(rotation="vertical")
plt.grid(True)
plt.show()
```

```python
g = sns.FacetGrid(data=df, row="brand", col="title_status", sharex=False, sharey=False, height=10)
g = g.map(sns.boxplot,"hours")
g.set_xticklabels(rotation="horizontal")
g.add_legend()
g.fig.set_figheight(100)
g.fig.set_figwidth(25)
g.fig.subplots_adjust(hspace=0.5)
plt.show()
```

```python
G = sns.PairGrid(data=df)
G.map_offdiag(sns.scatterplot)
G.map_diag(plt.hist)
G.fig.subplots_adjust(hspace=0.5)
for i in G.axes[4,:]:
    for j in i.get_xticklabels():
        j.set_rotation(55)
plt.show()
```
